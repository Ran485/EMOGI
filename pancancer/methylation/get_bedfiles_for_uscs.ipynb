```python
meth_df = pd.read_csv('../../data/pancancer/TCGA/methylation/download/002babbf-fae3-452e-b39b-e7e9823c48e6/jhu-usc.edu_BLCA.HumanMethylation450.24.lvl-3.TCGA-XF-AAMW-01A-11D-A42F-05.gdc_hg38.txt',
                      sep='\t')
meth_df.head()
```
